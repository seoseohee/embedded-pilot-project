```python
import cv2

def preprocess_terminal(img_bgr, scale=2, thr=150):
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    gray = cv2.resize(gray, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    _, binary = cv2.threshold(gray, thr, 255, cv2.THRESH_BINARY)
    return binary
```

```python
import easyocr, pytesseract
from PIL import Image
from transformers import TrOCRProcessor, VisionEncoderDecoderModel

reader = easyocr.Reader(["en"], gpu=False)
tess_config = "--oem 1 --psm 6"

processor = TrOCRProcessor.from_pretrained("microsoft/trocr-base-printed")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-base-printed")
model.eval()
```

```python
import subprocess, re, textwrap

def ocr_paddle_cli(img_path):
    # --use_angle_cls false / lang en
    cmd = [
        "paddleocr",
        "--image_dir", img_path,
        "--lang", "en",
        "--use_angle_cls", "false",
        "--show_log", "false",
    ]
    p = subprocess.run(cmd, capture_output=True, text=True)
    out = p.stdout + "\n" + p.stderr

    # paddleocr CLI 출력에서 인식 텍스트만 추출(형식이 버전마다 달라서 느슨하게)
    # 흔히 "text: xxx" 또는 "xxx\t0.99" 같은 형태가 섞임
    lines = []
    for line in out.splitlines():
        # text: ...
        m = re.search(r"text:\s*(.+)$", line)
        if m:
            lines.append(m.group(1).strip())
            continue
    # fallback: 너무 안 나오면 원문 일부 반환
    if not lines:
        return "(no parsed text)\n" + "\n".join(out.splitlines()[-20:])
    return "\n".join(lines)
```

```python
import time
import numpy as np

p = paths[0]
img = cv2.imread(p)
binimg = preprocess_terminal(img)

# EasyOCR
t0=time.time()
easy = "\n".join(reader.readtext(p, detail=0, paragraph=False))
t_easy=time.time()-t0

# Tesseract (전처리된 binary 사용)
t0=time.time()
tess = pytesseract.image_to_string(binimg, lang="eng", config=tess_config)
t_tess=time.time()-t0

# TrOCR (원본 PIL)
t0=time.time()
pil = Image.open(p).convert("RGB")
pixel_values = processor(images=pil, return_tensors="pt").pixel_values
generated_ids = model.generate(pixel_values)
trocr = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
t_trocr=time.time()-t0

# PaddleOCR (CLI)
t0=time.time()
paddle_txt = ocr_paddle_cli(p)
t_paddle=time.time()-t0

print("=== PaddleOCR (CLI) ===")
print(paddle_txt)
print(f"(time {t_paddle:.2f}s)")

print("\n=== EasyOCR ===")
print(easy)
print(f"(time {t_easy:.2f}s)")

print("\n=== Tesseract ===")
print(tess)
print(f"(time {t_tess:.2f}s)")

print("\n=== TrOCR ===")
print(trocr)
print(f"(time {t_trocr:.2f}s)")
```
